```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 30, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);
```

```python
print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)
```

```python
degree = 1
```

```python
coeffs = np.polyfit(x, y, degree)
```

```python
y_fit = np.polyval(coeffs, x)
```

```python
corr_matrix = np.corrcoef(y, y_fit)
```

```python
r_squared = corr_matrix[0, 1]**2
```

```python
print(f'The R-squared value is {r_squared:.5f}')
```

```python
import scipy.stats as stats
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
```

```python
result = stats.linregress(x, y)
plt.xlabel('Crude Oil Price', fontweight='bold')
plt.ylabel('GDP', fontweight='bold')
plt.scatter(x, y), plt.plot(x, result.intercept + result.slope*x,'r')
plt.title("Regression Output GDP on Crude Oil Price", fontsize=15, fontweight='bold')
plt.suptitle("Slope = -0.0127x, Intercept = 4.2828, R² = 0.0038, p-value = 0.6268, std.error = 0.0259, Y t-value = 0.627, X t-value = 0.011", fontsize=6.5, x=0.5, y=0.99, fontweight='bold')
plt.savefig('plot.png', dpi=900)
plt.show()
```
